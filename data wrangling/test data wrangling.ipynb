```python
#d)
nd = df.rename(columns = {'instant':'rec_id',
                                'dteday':'datetime',
                                'holiday':'is_holiday',
                                'workingday':'is_workingday',
                                'weathersit':'weather_condition',
                                'hum':'humidity',
                                'mnth':'month',
                                'cnt':'total_count',
                                'yr':'year'})
nd[:10]
```

```python
#e)
nd['season'] = nd['season'].astype('category') 
nd['holiday'] = nd['is_holiday'].astype('category') 
nd['workingday'] = nd['is_workingday'].astype('category') 
nd['weekday'] = nd['weekday'].astype('category') 
nd['weather condition'] = nd['weather_condition'].astype('category') 
nd['month'] = nd['month'].astype('category') 
nd['year'] = nd['year'].astype('category')
nd.dtypes
```
