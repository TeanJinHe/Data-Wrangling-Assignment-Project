```python
#Import the libraries that had been downloaded
import pandas as pd
import numpy as np
import sklearn
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
#Load the data 
vegetation_data = pd.read_csv("Vegetation_-_Central_Valley_Flood_Protection_Planning_Area_Update_-_2016.csv")
vegetation_data
```

```python
#Get information of data 
vegetation_data.info()
```

```python
#Check missing values using missingno library using matrix
msno.matrix(vegetation_data)
```

```python
#change fromdata to data frame when dropping the unused columns or too many missing data that does not affect the 
vege_df = vegetation_data.drop(['PER_CONIFE','ISOLATED_T','RESTORATIO','Comments','Source','MCVAlliance','CaCode','CommunityLink'],axis=1)
vege_df
```

```python
vege_df = vege_df[vege_df['GlobalRank'].notna()]
vege_df
```

```python
vege_df = vege_df[vege_df['SIZE_CATEG'].notna()]
vege_df
```

```python
vege_df = vege_df.replace(0, np.nan)
vege_df = vege_df.dropna(how='all', axis=0)
vege_df = vege_df.replace(np.nan, 0)
```

```python
#check again for any null values for all the columns
vege_df.info()
```

```python
#check for any duplicated data using boolean series 
bool_series = vege_df.duplicated()
bool_series
```

```python
#barplot for CalVegType and Acres 
sns.barplot(y = 'CalVegType',x = 'Acres',data = vege_df)
plt.xticks(rotation=90);
plt.show()
```

```python
#boxplot for density of plant per tree 
sns.boxplot(vege_df['PER_TREE'],color="skyblue")
```

```python
#boxplot for density of plant per hardwood 
sns.boxplot(vege_df['PER_HARDWO'],color="cyan")
```

```python
#boxplot for density of plant per shrub 
sns.boxplot(vege_df['PER_SHRUB'],color="magenta")
```

```python
#boxplot for density of plant per total 
sns.boxplot(vege_df['PER_TOTAL'],color="green")
```

```python
#sorted plot for global rank of each CV Group 
plt.subplot(121) ## one row and two column figures
df_sorted = vege_df.sort_values('GlobalRank')
plt.bar('GlobalRank','CV_Group', data=df_sorted, color='green')
plt.show()
```

```python
#heatmap for staterank for MCV Name for plants
sns.heatmap(pd.crosstab(vege_df.MCVName, vege_df.StateRank), cmap="Blues")
plt.show()
```

```python
#scatter plot for Shape length and Shape Area where it is differentiated using CWHR code 
fig = plt.figure()
ax = fig.add_subplot(111)
ax = sns.scatterplot(x='Shape__Length', y='Shape__Area',
hue='CWHRCode', data=vege_df)
plt.show()
```

```python
#line graph for size category and per shrub with category of CWHR Code 
sns.lineplot(data=vege_df, x="SIZE_CATEG", y="PER_SHRUB", hue="CWHRCode")
plt.xticks(rotation=90);
```

```python
#barplot of whether there are any distubrance in the clearing distance according to MCVLecel per tree
sns.barplot(data=vege_df, x='MCVLevel', y='PER_TREE', hue='CLEARING_D')
plt.xticks(rotation=90);
plt.show()
```
